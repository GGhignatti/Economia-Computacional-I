```python
# Programa soma002
# Descrição: Este programa lê dois números inteiros digitados pelo usuário e calcula a soma deles.
# Autor: Gabriel G. Ghignatti
# Data: 17/03/2026
# Versão: 0.0.3
# Notas da versão: Correção do erro semântico no uso da função input, a qual produzia a concatenação de textos ao em vez de soma de números.
```

```python
# Alocação de Memória

numero_1 = 0

numero_2 = 0

soma = 0
```

```python
# Entrada de Dados
numero_1 = int(input("Qual a primeira parcela: ")) # Acréscimo da função int()

numero_2 = int(input("Qual o segunda parcela: ")) # Acréscimo da função int()
```

```python
# Processamento de Dados: Cálculo da Soma

soma = numero_1 + numero_2
```

```python
# Saída de Dados
print(soma)
```
